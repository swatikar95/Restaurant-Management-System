```python
from tkinter import *
import random
import time

window=Tk()
window.geometry('1600x800')
window.title("Restaurant Managment System")

Tops = Frame(window,width=1600,height=50,bg='powder blue')
Tops.pack(side=TOP)

f1 = Frame(window,width=800,height=700)
f1.pack(side=LEFT)

f2 = Frame(window,width=300,height=700)
f2.pack(side=RIGHT)
#time
localtime=time.asctime(time.localtime(time.time()))
#Info
lblInfo=Label(Tops,font=('arial',50,'bold'),text="Restaurant Managment System",fg='Steel Blue')
lblInfo.grid(row=0,column=0)
lblInfo=Label(Tops,font=('arial',20,'bold'),text=localtime,fg='Steel Blue')
lblInfo.grid(row=1,column=0)
#calculator
text_input=StringVar()
expression = ""

def press(numbers):
    global expression
    expression = expression + str(numbers)
    text_input.set(expression)
def clear():
    global expression
    expresion=""
    text_input.set("")

def equalTo():
    global expression
    total=str(eval(expression))
    text_input.set(total)
    expression=""
    
txtDisplay = Entry(f2,font=('arial',20),bd=10,textvariable=text_input,bg='powder blue',insertwidth=4,justify='right')
txtDisplay.grid(columnspan=8)

btn7=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='7',bg='powder blue',command=lambda:press(7))
btn7.grid(row=2,column=0)
btn8=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='8',bg='powder blue',command=lambda:press(8))
btn8.grid(row=2,column=1)
btn9=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='9',bg='powder blue',command=lambda:press(9))
btn9.grid(row=2,column=2)

btn4=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='4',bg='powder blue',command=lambda:press(4))
btn4.grid(row=3,column=0)
btn5=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='5',bg='powder blue',command=lambda:press(5))
btn5.grid(row=3,column=1)
btn6=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='6',bg='powder blue',command=lambda:press(6))
btn6.grid(row=3,column=2)

btn1=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='1',bg='powder blue',command=lambda:press(1))
btn1.grid(row=4,column=0)
btn2=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='2',bg='powder blue',command=lambda:press(2))
btn2.grid(row=4,column=1)
btn3=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='3',bg='powder blue',command=lambda:press(3))
btn3.grid(row=4,column=2)

addition=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='+',bg='powder blue',command=lambda:press("+"))
addition.grid(row=2,column=3)
subtraction=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='-',bg='powder blue',command=lambda:press("-"))
subtraction.grid(row=3,column=3)
mul=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='*',bg='powder blue',command=lambda:press('*'))
mul.grid(row=4,column=3)

btn0=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='0',bg='powder blue',command=lambda:press(0))
btn0.grid(row=5,column=0)
clear=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='C',bg='powder blue',command=clear)
clear.grid(row=5,column=1)
equal=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='=',bg='powder blue',command=equalTo)
equal.grid(row=5,column=2)
division=Button(f2,padx=16,pady=16,fg='black',font=('arial',20),text='/',bg='powder blue',command=lambda:press('/'))
division.grid(row=5,column=3)
#end calculator

#restaurant info 1
rand=StringVar()
Fries=StringVar()
Burger=StringVar()
Filet=StringVar()
subTotal=StringVar()
Total=StringVar()
Services=StringVar()
Drinks=StringVar()
Tax=StringVar()
Cost=StringVar()
Chicken_Burger=StringVar()
Cheese_Burger=StringVar()

lblReference=Label(f1,font=('arial',16,'bold'),text='Reference')
lblReference.grid(row=0,column=0)
txtReference = Entry(f1,font=('arial',16,'bold'),textvariable=rand,bd=5,insertwidth=4,bg='powder blue',justify='right')
txtReference.grid(row=0,column=1)

lblFries=Label(f1,font=('arial',16,'bold'),text='Large Fries')
lblFries.grid(row=1,column=0)
txtFries = Entry(f1,font=('arial',16,'bold'),textvariable=Fries,bd=5,insertwidth=4,bg='powder blue',justify='right')
txtFries.grid(row=1,column=1)

lblBurger=Label(f1,font=('arial',16,'bold'),text='Burger Meal')
lblBurger.grid(row=2,column=0)
txtBurger = Entry(f1,font=('arial',16,'bold'),textvariable=Burger,bd=5,insertwidth=4,bg='powder blue',justify='right')
txtBurger.grid(row=2,column=1)

lblFilet=Label(f1,font=('arial',16,'bold'),text='Filet Meal')
lblFilet.grid(row=3,column=0)
txtFilet = Entry(f1,font=('arial',16,'bold'),textvariable=Filet,bd=5,insertwidth=4,bg='powder blue',justify='right')
txtFilet.grid(row=3,column=1)

lblChicken=Label(f1,font=('arial',16,'bold'),text='Chicken Burger')
lblChicken.grid(row=4,column=0)
txtChicken = Entry(f1,font=('arial',16,'bold'),textvariable=Chicken_Burger,bd=5,insertwidth=4,bg='powder blue',justify='right')
txtChicken.grid(row=4,column=1)

lblCheese=Label(f1,font=('arial',16,'bold'),text='Cheese Burger')
lblCheese.grid(row=5,column=0)
txtCheese = Entry(f1,font=('arial',16,'bold'),textvariable=Cheese_Burger,bd=5,insertwidth=4,bg='powder blue',justify='right')
txtCheese.grid(row=5,column=1)

#restaurant info 2
lblDrinks=Label(f1,font=('arial',16,'bold'),text='Drinks')
lblDrinks.grid(row=0,column=2)
txtDrinks = Entry(f1,font=('arial',16,'bold'),textvariable=Drinks,bd=5,insertwidth=4,bg='powder blue',justify='right')
txtDrinks.grid(row=0,column=3)

lblCost=Label(f1,font=('arial',16,'bold'),text='Cost of Meal')
lblCost.grid(row=1,column=2)
txtCost = Entry(f1,font=('arial',16,'bold'),textvariable=Cost,bd=5,insertwidth=4,bg='powder blue',justify='right')
txtCost.grid(row=1,column=3)

lblTax=Label(f1,font=('arial',16,'bold'),text='State Tax')
lblTax.grid(row=2,column=2)
txtTax = Entry(f1,font=('arial',16,'bold'),textvariable=Tax,bd=5,insertwidth=4,bg='powder blue',justify='right')
txtTax.grid(row=2,column=3)

lblServices=Label(f1,font=('arial',16,'bold'),text='Service Charge')
lblServices.grid(row=3,column=2)
txtServices = Entry(f1,font=('arial',16,'bold'),textvariable=Services,bd=5,insertwidth=4,bg='powder blue',justify='right')
txtServices.grid(row=3,column=3)

lblsubTotal=Label(f1,font=('arial',16,'bold'),text='Sub Total')
lblsubTotal.grid(row=4,column=2)
txtsubTotal = Entry(f1,font=('arial',16,'bold'),textvariable=subTotal,bd=5,insertwidth=4,bg='powder blue',justify='right')
txtsubTotal.grid(row=4,column=3)

lblTotal=Label(f1,font=('arial',16,'bold'),text='Total Cost')
lblTotal.grid(row=5,column=2)
txtTotal = Entry(f1,font=('arial',16,'bold'),textvariable=Total,bd=5,insertwidth=4,bg='powder blue',justify='right')
txtTotal.grid(row=5,column=3)

#button
def Ref():
    x=random.randint(10000,12000)
    randomRef = str(x)
    rand.set(randomRef)
    
    CoF = float(Fries.get())
    CoD = float(Drinks.get())
    CoFilet = float(Filet.get())
    CoBurger = float(Burger.get())
    CoChiBurger = float(Chicken_Burger.get())
    CoCheBurger = float(Cheese_Burger.get())
    
    CostofFries = CoF*0.99
    CostofDrinks = CoD*1.00
    CostofFilet = CoFilet*2.99
    CostofBurger = CoBurger*2.87
    CostofchiBurger = CoChiBurger*2.89
    CostofcheBurger = CoCheBurger*2.69
    
    CostofMeal = "$", str("%.2f" %(CostofFries+CostofDrinks+CostofFilet+
                                  CostofBurger+CostofchiBurger+CostofcheBurger))
    payTax= ((CostofFries+CostofDrinks+CostofFilet+
                                  CostofBurger+CostofchiBurger+CostofcheBurger)*.2)
    TotalCost = (CostofFries+CostofDrinks+CostofFilet+
                                  CostofBurger+CostofchiBurger+CostofcheBurger)
    ser_charge = ((CostofFries+CostofDrinks+CostofFilet+
                                  CostofBurger+CostofchiBurger+CostofcheBurger)/99)
    services = "$", str("%.2f" % ser_charge)
    
    overallCost = "$", str("%.2f" % (payTax+TotalCost+ser_charge))
    paidTax = "$", str("%.2f" % payTax)
    
    Services.set(services)
    Cost.set(CostofMeal)
    Tax.set(paidTax)
    subTotal.set(CostofMeal)
    Total.set(overallCost)

def Exit():
    window.destroy()

def Reset():
    rand.set("")
    Fries.set("")
    Burger.set("")
    Filet.set("")
    subTotal.set("")
    Total.set("")
    Services.set("")
    Drinks.set("")
    Tax.set("")
    Cost.set("")
    Chicken_Burger.set("")
    Cheese_Burger.set("")
    
btnTotal = Button(f1,padx=8,pady=8,fg='black',font=('arial',16,'bold'),width=10,text='Total',bg='powder blue',command=Ref)
btnTotal.grid(row=7,column=1)
btnExit = Button(f1,padx=8,pady=8,fg='black',font=('arial',16,'bold'),width=10,text='Exit',bg='powder blue',command=Exit)
btnExit.grid(row=7,column=2)
btnReset = Button(f1,padx=8,pady=8,fg='black',font=('arial',16,'bold'),width=10,text='Reset',bg='powder blue',command=Reset)
btnReset.grid(row=7,column=3)

window.mainloop()
```
